### Lets try to find the most frequently occurring position of each letter

```python
letter_pos1 = dict( (key, 0) for key in string.ascii_lowercase )
letter_pos2 = dict( (key, 0) for key in string.ascii_lowercase )
letter_pos3 = dict( (key, 0) for key in string.ascii_lowercase )
letter_pos4 = dict( (key, 0) for key in string.ascii_lowercase )
letter_pos5 = dict( (key, 0) for key in string.ascii_lowercase )
### letter_posx denotes the no of times the letter occurs in position x
```

```python
for i in wordle_list:
    letter_pos1[i[0]]+=1
    letter_pos2[i[1]]+=1
    letter_pos3[i[2]]+=1
    letter_pos4[i[3]]+=1
    letter_pos5[i[4]]+=1
```

```python
letter_pos1
```

```python
def pos_graph_func(a):
    ### function to plot the graph depicting the frequency distribution of letter in words
    x=[letter_pos1[a],letter_pos2[a],letter_pos3[a],letter_pos4[a],letter_pos5[a]]
    y=[1,2,3,4,5]
    plt1 = plt.figure(figsize=(12, 5))
    plt.bar(y,x,width=0.9,align='center',color='#006400',edgecolor="#004b23")
#     fig = plt.figure(figsize = (35, 20))
    
    plt.xlabel("Position of Letter")
    plt.ylabel("No. of times it occurs at given position")
    plt.title("Distribution of the letter '"+a+"' across Wordle words")
    plt.show()
    
```

```python
# running the pos_graph_func() for a few letters
pos_graph_func('a')
```

```python
max_pos=dict( (key, 0) for key in string.ascii_lowercase )
for i in string.ascii_lowercase:
    t=max(letter_pos1[i],letter_pos2[i],letter_pos3[i],letter_pos3[i],letter_pos4[i])
#     print(t)
    if(letter_pos1[i]==t):
        max_pos[i]=1
    elif(letter_pos2[i]==t):
        max_pos[i]=2
    elif(letter_pos3[i]==t):
        max_pos[i]=3
    elif(letter_pos4[i]==t):
        max_pos[i]=4
    elif(letter_pos5[i]==t):
        max_pos[i]=5
        
max_pos
### max_pos denotes the position in which the letter occurs the most
```

### Now let's find the most frequently occurring letters in Wordle words with one letter repeating

```python
wordle4=[]
for i in wordle_list:
    a=set(i)
    if(len(a)==4):
        wordle4.append(i)
        
w4 = dict( (key, 0) for key in string.ascii_lowercase )


for i in wordle4:
    for j in i:
        if(i.count(j)>=2):
            w4[j]+=1

# w4
w4=dict(sorted(w4.items(), key=lambda item: item[1],reverse=True))
w4
            
        
        
```

### If the Wordle Word Of The Day has only 4 unique characters, it is most likely that the repeating letter is either e, a, o, t, r, l, s or n

### It is also interesting to note that the letters q and x will never repeat and letters w and j have very little likelihood of repeating

```python
### convert the letter_pos x values into a numpy array 
### call the sns heatmap method
col=[]

for i in string.ascii_lowercase:
    sam=[letter_pos1[i],letter_pos2[i],letter_pos3[i],letter_pos4[i],letter_pos5[i]]
    col.append(sam)

# letter_pos1[0]
# col
y=[]
for i in string.ascii_lowercase:
    y.append(i)
letter_pos_data = pd.DataFrame(col, columns =['First', 'Second', 'Third','Fourth','Fifth'],index=y)
letter_pos_data
```

```python
import seaborn as sns
sns.set_theme()
fig, ax = plt.subplots(figsize=(20,10)) 
ax = sns.heatmap(letter_pos_data,linecolor='black',linewidths=0.0002,cmap="Blues",cbar=True)
ax.set_title("Heatmap Showing Frequency of Each Letter at Each Position")
# plt.pcolor(letter_pos_data)
# plt.show()
```
